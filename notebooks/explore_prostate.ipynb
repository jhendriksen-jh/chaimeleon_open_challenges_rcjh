```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_file = '/home/jordan/github/chaimeleon_open_challenges_rcjh/datasets/train_prostate/train/case_0086/case_0086.nii.gz'
```

```python
nifti_image = nib.load(image_file)
nii_data = nifti_image.get_fdata()
```

```python
nii_chunks = []
for k in range(3):
    u = int(np.ceil((k+1)*(nii_data.shape[-1]/3)))
    l = int(np.floor((k)*(nii_data.shape[-1]/3)))
    nii_chunk = np.mean(nii_data[:,:,l:u], axis=2)
    nii_chunk = ((nii_chunk - nii_chunk.min()) / (nii_chunk.max() - nii_chunk.min()))*255
    nii_chunks.append(nii_chunk)
nii_chunked_image = np.array(nii_chunks)
nii_chunked_image = np.transpose(nii_chunked_image, (1,2,0))
```

```python
fig, ax = plt.subplots(figsize=(20,15), ncols=3)
ax[0].imshow(nii_chunked_image[:,:, 0])
ax[1].imshow(nii_chunked_image[:,:,1])
ax[2].imshow(nii_chunked_image[:,:,2])
```
